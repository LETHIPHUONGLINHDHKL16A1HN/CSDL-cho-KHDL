```python
# INNER JOIN
df_inner = df_student.merge(df_course, how="inner", left_on="course_id", right_on="id")
print("üîπ INNER JOIN:")
df_inner
```

#
 LEFT JOIN: L·∫•y t·∫•t c·∫£ sinh vi√™n, k·ªÉ c·∫£ nh·ªØng ng∆∞·ªùi ch∆∞a c√≥ m√¥n h·ªçc h·ª£p l·ªá (course_id kh√¥ng t·ªìn t·∫°i trong course)

SELECT student.*, course.course_name
FROM student
LEFT JOIN course ON student.course_id = course.id;

```python
# LEFT JOIN
df_left = df_student.merge(df_course, how="left", left_on="course_id", right_on="id")
print("üîπ LEFT JOIN:")
df_left
```

#
 RIGHT JOIN: L·∫•y t·∫•t c·∫£ m√¥n h·ªçc, k·ªÉ c·∫£ nh·ªØng m√¥n kh√¥ng c√≥ sinh vi√™n h·ªçc

SELECT student.*, course.course_name
FROM student
RIGHT JOIN course ON student.course_id = course.id;


```python
df_right = df_course.merge(df_student, how="right", left_on="id", right_on="course_id")

print("üîπ RIGHT JOIN:")
display(df_right[["student_id", "name", "class", "course_name", "score"]])
```

```python
# FULL OUTER JOIN: K·∫øt h·ª£p c·∫£ hai b·∫£ng
df_full_outer = pd.merge(df_student, df_course, how="outer", left_on="course_id", right_on="id")

print("üîπ FULL OUTER JOIN (K·∫øt h·ª£p c·∫£ hai b·∫£ng):")
display(df_full_outer[["student_id", "name", "class", "course_name", "score"]])
```

FROM student

```python
df_class_stats = df_student.groupby("class").agg(
    total_students=("student_id", "count"),
    avg_score=("score", "mean")
).reset_index()

print("üîπ T·ªïng s·ªë sinh vi√™n v√† ƒëi·ªÉm trung b√¨nh t·ª´ng l·ªõp:")
df_class_stats
```

# Ph√¢n lo·∫°i sinh vi√™n theo ƒëi·ªÉm s·ªë

# ph√¢n lo·∫°i sinh vi√™n 
SELECT name, score,
    CASE
        WHEN score >= 9.0 THEN 'Xu·∫•t s·∫Øc'
        WHEN score >= 6.0 THEN 'T·ªët'
        ELSE 'K√©m'
    END AS rank_category
FROM student;


```python
def classify_score(score):
    if score >= 9.0:
        return "Xu·∫•t s·∫Øc"
    elif score >= 6.0:
        return "T·ªët"
    else:
        return "K√©m"

df_student["rank_category"] = df_student["score"].apply(classify_score)

print("üîπ Ph√¢n lo·∫°i sinh vi√™n theo ƒëi·ªÉm:")
df_student[["name", "score", "rank_category"]]
```


---- theo m√¥n h·ªçc

```python
# G·ªôp d·ªØ li·ªáu sinh vi√™n v·ªõi t√™n m√¥n h·ªçc
df_rank_by_course = df_student.merge(df_course, how="left", left_on="course_id", right_on="id")

# X·∫øp h·∫°ng sinh vi√™n theo m√¥n h·ªçc d·ª±a tr√™n ƒëi·ªÉm
df_rank_by_course["rank_in_course"] = df_rank_by_course.groupby("course_name")["score"].rank(method="dense", ascending=False)

print("üîπ X·∫øp h·∫°ng sinh vi√™n theo m√¥n h·ªçc:")
display(df_rank_by_course.sort_values(by=["course_name", "rank_in_course"]))
```

# Th√™m c·ªôt graduation_date v√†o b·∫£ng student v√† c·∫≠p nh·∫≠t d·ª±a tr√™n ƒëi·ªÉm s·ªë
ALTER TABLE student ADD COLUMN graduation_date DATETIME;
UPDATE student
SET graduation_date = 
    CASE
        WHEN score >= 9.0 THEN DATE_ADD(NOW(), INTERVAL 1 YEAR)
        WHEN score >= 6.0 THEN DATE_ADD(NOW(), INTERVAL 4 YEAR)
        ELSE DATE_ADD(NOW(), INTERVAL 5 YEAR)
    END;

```python
def calculate_graduation_date(score):
    if score >= 9.0:
        return datetime.now() + timedelta(days=365)  # 1 nƒÉm
    elif score >= 6.0:
        return datetime.now() + timedelta(days=4*365)  # 4 nƒÉm
    else:
        return datetime.now() + timedelta(days=5*365)  # 5 nƒÉm

df_student["graduation_date"] = df_student["score"].apply(calculate_graduation_date)

print("‚úÖ ƒê√£ c·∫≠p nh·∫≠t graduation_date d·ª±a tr√™n ƒëi·ªÉm s·ªë.")
df_student[["name", "score", "graduation_date"]]
```
